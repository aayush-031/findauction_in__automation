```python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
from tqdm import tqdm
from pprint import pprint
import pickle
```

```python
def clean_text(data):
    data= ' '.join(data.split())
    return data
```

```python
# url ='https://findauction.in/bank-property/mumbai/all/all/430'
# url.split('/')[4]
```

```python
def scrapweb(obj1,driver):
    driver.get(obj1)
    path=(driver.current_url)
#     print(path)
    path_new = (path +"/all/all/" +str(5000))
#     driver.close()
#     driver = webdriver.Safari()
    driver.get(path_new)
    path2=driver.current_url
    city=path2.split('/')[4]
    n= int(path2.split('/')[-1])
#     driver.close()
#     driver = webdriver.Safari()
    dfs =pd.DataFrame()
    for i in range(1,n):
        driver.get(path+"/all/all/"+str(i))
 #   print(path2)
        obj_dic = {}
        pro_details=driver.find_elements_by_class_name("property-list-detail")
        for obj in tqdm(pro_details):
            df = pd.DataFrame()
            title=obj.find_element_by_tag_name("a").text
            url=obj.find_element_by_tag_name("a").get_attribute("href")
            ##print(url)
            title= clean_text(title)
            ##print(title)
            bank=obj.find_element_by_tag_name("small").text
            bank =clean_text(bank)
            ##print(bank)
            price=obj.find_element_by_tag_name("h6").text.replace(',','')
            price=int(price)
            ##print(price)
            li_ = ""
            un_list=obj.find_element_by_tag_name("ul").find_elements_by_tag_name("li")
            for li_list in un_list:
                li_ += "".join(li_list.text.split())
                li_ += "|"
            extra_info=li_
            obj_dic[url] = {
                "Title" : title, 
                 "url":url,
                "Bank": bank,
                 "price":price,
                "Extra_Info":extra_info 
                }
            df=pd.DataFrame(obj_dic).T
#             driver.close()
            dfs=dfs.append(df)
    dfs['City'] =city
    return dfs
        
            ##un_list= "|".join(un_list.split())
            ##print(li_)
```

```python
df1 = pd.DataFrame()
driver = webdriver.Chrome()
driver.get('https://findauction.in/')
# driver.implicitly_wait(5)
soup = BeautifulSoup(driver.page_source,'lxml')
listobjlinks = driver.find_element_by_id("tab1").find_elements_by_tag_name("a")
urls_tab1 =[_.get_attribute("href") for _ in listobjlinks]
for obj in tqdm(urls_tab1):
    df1=df1.append(scrapweb(obj,driver))
print(df1.head())
driver.close()
type(listobjlinks)
```

```python
df1.to_pickle("Find_Auction_top_cities.pkl")
```
